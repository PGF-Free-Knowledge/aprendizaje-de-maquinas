```python
# =====================================================================
# 🔍 BUSCADOR DETALLADO + SUBTOTALES POR MODELO
# =====================================================================

import pandas as pd

# Palabra clave a buscar en la descripción general
palabra_clave = "Tarjeta de desarrollo"   # Ejemplo: "osciloscopio", "fuente", "analizador"

# Filtrar filas que contengan la palabra clave en la columna de descripción
filtro = df['Descripción del Activo Fijo'].str.contains(palabra_clave, case=False, na=False)
resultado = df[filtro]

print(f"🔹 Instrumentos que contienen '{palabra_clave}' en la descripción: {len(resultado)} encontrados\n")

# Agrupar por las columnas relevantes incluyendo el estado
agrupado = (
    resultado.groupby([
        'Descripción del Activo Fijo',
        'Marca',
        'Modelo',
        'Instrumento (Especificación)',
        'Estado del Activo Fijo'
    ])
    .size()
    .reset_index(name='Cantidad')
    .sort_values(by=['Descripción del Activo Fijo', 'Marca', 'Modelo'], ascending=True)
)

# === Agregar subtotales por modelo ===
subtotales = (
    agrupado.groupby(['Descripción del Activo Fijo', 'Marca', 'Modelo'], as_index=False)['Cantidad']
    .sum()
)

# Marcar los subtotales con una etiqueta especial
subtotales['Instrumento (Especificación)'] = '--- SUBTOTAL ---'
subtotales['Estado del Activo Fijo'] = ''
# Unir con el detalle completo
agrupado_con_subtotales = pd.concat([agrupado, subtotales], ignore_index=True)
agrupado_con_subtotales = agrupado_con_subtotales.sort_values(
    by=['Descripción del Activo Fijo', 'Marca', 'Modelo', 'Instrumento (Especificación)']
).reset_index(drop=True)

# Mostrar los primeros resultados
display(agrupado_con_subtotales.head(40))

# (Opcional) Guardar el resultado detallado con subtotales
# agrupado_con_subtotales.to_csv(f"inventario_{palabra_clave}_detallado_subtotales.csv", index=False)
```
